**Question 1:** What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly explain with examples.

**Answer:**

**Paste**
* Definition: Pastes everything (values, formulas, formatting, comments, etc.) from the copied cells.
* Use case: When you want to duplicate the entire content of a cell or range.

Example:
* Cell A1 contains =B1+C1 and is formatted bold.
* If you copy A1 and use Paste in A2:
  * A2 will have the formula =B2+C2 and also be bold.

**Paste Special**
* Definition: Gives you control over what to paste (e.g., only values, only formatting, only formulas, etc.).
* Use case: When you want to paste specific elements of the copied data.

Examples:
1. Paste Special → Values
    * If A1 has =B1+C1 and the result is 10, using Paste Special > Values in A2 will paste just the number 10, not the formula.
2. Paste Special → Formats
    * Copies only the formatting (e.g., bold, color) without changing the content of the target cells.
3. Paste Special → Transpose
    * Converts rows to columns or vice versa when pasting.



**Question 2:** Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes' in Excel.

**Answer:**

**Freeze Panes**
  
* **Function:**
  * Locks rows and/or columns so they remain visible while you scroll through the rest of the worksheet.

* **Usefulness:**
  * Helps keep headers or labels in view while working with large datasets.
  * Improves navigation and data interpretation.

* **Examples:**
1. Freeze Top Row:
    * Keeps the top row (e.g., column headers) visible as you scroll down.

2. Freeze First Column:
    * Keeps the first column (e.g., row labels like names or IDs) visible when scrolling horizontally.

3. Freeze Panes (custom):
    * You can freeze rows above and columns to the left of the selected cell.
    * Example: Selecting cell B2 and choosing "Freeze Panes" will freeze Row 1 and Column A.

**Split Panes**
* **Function:**
    * Divides the worksheet into two or four independent scrollable panes.

* **Usefulness:**
    * Allows viewing different parts of the same worksheet at the same time.
    * Helpful for comparing data from distant parts of a large sheet side by side or top and bottom.

* **Examples:**

1. Split Horizontally:
    * Splits into upper and lower sections you can scroll independently.

2. Split Vertically:
    * Splits into left and right panes for side-by-side comparison.

3. Split Both Ways:
    * Selecting a cell (e.g., C5) and clicking "Split" will divide the screen into four panes around that cell.

| Feature         | Freeze Panes                       | Split Panes                              |
| --------------- | ---------------------------------- | ---------------------------------------- |
| Purpose         | Keep headers/labels visible        | View multiple parts of a sheet at once   |
| Scroll Behavior | Locked rows/columns                | Independent scrolling areas              |
| Visibility      | Keeps selected part **fixed**      | No part is fixed—just scrolls separately |
| Use Case        | Data entry, reviewing large tables | Comparing distant data on same sheet     |


**Question 3:** Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?

***Answer:***

**Insert a New Row or Column**
* **To insert a row:**
  1. Right-click the row number where you want the new row.
  2. Select Insert.
  3. A new row appears above the selected row.
* **To insert a column:**
  1. Right-click the column letter where you want the new column.
  2. Select Insert.
  3. A new column appears to the left of the selected column.

**Insert Multiple Rows or Columns at Once**
* **Insert multiple rows:**
  * Select the same number of existing rows as you want to insert (e.g., highlight 3 rows).
  * Right-click and choose Insert.
  * Excel inserts 3 new rows above the selection.

* **Insert multiple columns:**
  * Select the same number of existing columns (e.g., highlight columns B to D to insert 3).
  * Right-click and choose Insert.
  * Excel inserts 3 new columns to the left of the selection.

| Feature                    | Inserting a Row                                    | Inserting a Column                                |
| -------------------------- | -------------------------------------------------- | ------------------------------------------------- |
| **Direction of Insertion** | Inserts a horizontal line of cells (left to right) | Inserts a vertical line of cells (top to bottom)  |
| **Location Shift**         | Existing rows below shift **down**                 | Existing columns to the right shift **right**     |
| **Use Case**               | Add new data records (e.g., new entries)           | Add new data fields (e.g., new categories)        |
| **Reference Impact**       | Changes row numbers (e.g., row 5 becomes row 6)    | Changes column letters (e.g., column C becomes D) |


**Question 4:** What are logical functions in Excel? Provide examples of at least two logical functions and their applications.

**Answer:**

Logical functions in Excel are used to perform comparisons, tests, and return values based on whether a condition is TRUE or FALSE. These are essential for decision-making tasks like filtering, flagging data, and automating calculations.

**Logical Functions with Examples**

**1. IF Function**

**Purpose:**
Returns one value if a condition is TRUE, and another if the condition is FALSE.

Syntax:

    =IF(logical_test, value_if_true, value_if_false)
Example:

    =IF(A2>=60, "Pass", "Fail")
  
  * Application: Used to determine if a student passed based on their score in cell A2.
  * If the score is 60 or higher, it returns "Pass"; otherwise, "Fail".

**2. AND Function**

**Purpose:** Returns TRUE if all conditions are TRUE, otherwise returns FALSE.

Syntax:

    =AND(condition1, condition2, ...)
Example:

    =AND(A2>=60, B2="Yes")

  * Application: Used to check if a student passed (score ≥ 60) and attended class.

  * Returns TRUE only if both conditions are met.

**Other Useful Logical Functions**
  * OR() – Returns TRUE if any condition is true.
  * NOT() – Reverses the result (e.g., NOT(TRUE) returns FALSE).
  * IFERROR() – Handles errors gracefully (e.g., display “Invalid” if a formula errors).

**Question 5:** Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.

**Answer:**

XLOOKUP is a powerful lookup function introduced in Excel 365 and Excel 2019+. It's designed to replace older lookup functions like VLOOKUP, HLOOKUP, and LOOKUP with a more flexible, reliable, and easier-to-read syntax.

**Purpose:**
XLOOKUP searches a range or array for a specified value and returns a corresponding value from another range or array.

**Syntax of XLOOKUP**

    =XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found], [match_mode], [search_mode])

* lookup_value: The value you want to search for.
* lookup_array: The column or row to search.
* return_array: The column or row from which to return the result.
* [if_not_found]: (Optional) What to return if no match is found.
* [match_mode]: (Optional) 0 for exact match (default), -1 or 1 for next smaller/larger match.
* [search_mode]: (Optional) 1 for first-to-last (default), -1 for last-to-first.

**Example Comparison**

**XLOOKUP:**

    =XLOOKUP("John", A2:A10, B2:B10, "Not Found")

* Looks for "John" in column A, returns value from column B.
* If not found, shows "Not Found".

**VLOOKUP:**

    =IFERROR(VLOOKUP("John", A2:B10, 2, FALSE), "Not Found")

* More complex and less flexible.

** Why Use XLOOKUP**
* Easier to read and maintain
* Works in both directions
* Reduces chances of errors (no need to count columns)
* Offers more robust error handling
* Supports dynamic arrays

| Feature                    | `XLOOKUP`                           | `VLOOKUP`                            |
| -------------------------- | ----------------------------------- | ------------------------------------ |
| **Search Direction**       | Vertical **or** horizontal          | Only vertical (column-based)         |
| **Return Direction**       | Can return values **left or right** | Only returns values **to the right** |
| **Exact Match Default**    | Yes (default is exact match)        | No (default is approximate match)    |
| **Column Index**           | Not needed (uses array references)  | Requires hardcoded index number      |
| **Error Handling**         | Built-in `[if_not_found]` option    | Requires `IFERROR()` separately      |
| **Dynamic Arrays**         | Supports returning arrays           | Does not                             |
| **Backward Compatibility** | Newer Excel only (365/2019+)        | Works in all Excel versions          |

**Question 10:** You're working with a dataset that contains missing values. As a Data Scientist, explain how you'd detect and handle missing data using Excel.

Mention tools like:
* Go To Special
* ISBLANK
* COUNTBLANK

(Include a screenshot showing how blanks are identified or processed.)

**Answer:**

**Use ISBLANK Function**

**Purpose:** Checks if a specific cell is blank.

Formula Example:

    =ISBLANK(B2)
This returns:
* TRUE if B2 is empty
* FALSE if it contains data

**Use COUNTBLANK Function**

**Purpose:** Counts the number of blank cells in a range.

Formula Example:

    =COUNTBLANK(B2:B10)

* Returns the number of missing values in the selected range.
